```python
import numpy as np
import matplotlib.pyplot as plt

def generate_dataset(d):
    return np.random.uniform(low=-1, high=1, size=(1000, d))
    
def closest_point_distance(dataset):
    distances_to_origin = np.linalg.norm(dataset, axis=1)
    return np.min(distances_to_origin)

dimensions = [2, 4, 8, 16, 32, 64]
closest_distances = []
for d in dimensions:
    dataset = generate_dataset(d)
    closest_distance = closest_point_distance(dataset)
    closest_distances.append(closest_distance)

# plot resuluts
plt.plot(dimensions, closest_distances, marker='o')
plt.xlabel('Dimension (d)')
plt.ylabel('Closest Point Distance to Origin')
plt.title('Curse of Dimensionality - Closest Point to Origin')
plt.show()
```

```python
def point_distances(dataset):
    distances_to_origin = np.linalg.norm(dataset, axis=1)
    closest_distance = np.min(distances_to_origin)
    furthest_distance = np.max(distances_to_origin)
    return closest_distance, furthest_distance

ratio_distances = []
for d in dimensions:
    dataset = generate_dataset(d)
    closest_distance, furthest_distance = point_distances(dataset)
    ratio_distances.append(furthest_distance / closest_distance)

# Plot results
plt.plot(dimensions, ratio_distances, marker='o')
plt.xlabel('Dimension (d)')
plt.ylabel('Ratio ∆1(d) / ∆0(d)')
plt.title('Curse of Dimensionality - Ratio of Furthest to Closest Point')
plt.show()
```
